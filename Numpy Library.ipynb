```python
import numpy as np
```

```python
# vector
# [1, 2, 3]
a = np.array([1,2,3])
print(a.shape)
a[0] = 5
print(a[0])

# matrix
# [[1,2,3], [3,4,5]]
b = np.array([[1,2,3], [4,5,6]])
print(b.shape)

# tensor
c = np.array([[[1,2,3], [4,5,6]],
            [[1,2,3], [4,5,6]],
            [[1,2,3], [4,5,6]],
            [[1,2,3], [4,5,6]]])
print(c.shape)
```

ARRAY INDEXING

```python
a = np.arange(1,13).reshape(3,4)
```

```python
print(a)
```

```python
b = a[:, 2:]
print(b)
```

```python
b = a[:, :2]
print(b)
```

```python
b = np.copy(a[:-2, :1])
print(b)
```

```python
b[0,0] = 10
print(a)
```
