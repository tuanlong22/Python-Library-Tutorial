```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph


```python
x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]

#Resize the graph
plt.figure(figsize = (6,4), dpi =100)

plt.plot(x,y, label = "2x", color = "y", linewidth = 2, marker = '.',linestyle = '--', markersize = 10, markeredgecolor = 'pink')
# Line number two
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, label = 'x^2', linestyle = '-')
plt.plot(x2[5:],x2[5:]**2, linestyle='--')

plt.title("Our first graph",fontdict = {'fontname' : 'Comic Sans MS', 'fontsize':20})
plt.xlabel('X axis',color = 'red')
plt.ylabel('Y axis', color = "green")
plt.xticks([0,1,2,3,4,5])
#plt.yticks([0,2,4,6,8])
plt.legend()
plt.savefig('mygraph.png',dpi = 300)
plt.show()
```

# Bar chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]
bars = plt.bar(labels,values)
patterns = ['/','O','*']
for bar in bars:
  bar.set_hatch(patterns.pop(0))
bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')
plt.figure(figsize = (6,4))
plt.show()
```

# Real World Example


Line Graph


```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize = (8,5))
plt.title('Gas prices over Time (in USD)')
country_to_look_at = ['Australia','USA','South Korea','Canada']
for country in gas:
  if country  in country_to_look_at:
    plt.plot(gas.Year, gas[country], marker = '.')
# plt.plot(gas.Year, gas.USA,'b.-', label = "United States")
# plt.plot(gas.Year, gas.Canada,'r.-', label = "Canada")
# plt.plot(gas.Year, gas['South Korea'],'g.-', label = "South Korea")
# plt.plot(gas.Year, gas.Australia, 'y.-', label = "Australia")
# print(gas.Year[::3])
plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('US Dollar')
plt.legend()
plt.savefig('Gas_prices_over_time.png', dpi = 300)
plt.show()
```

LOAD FIFA DATA

```python
fifa = pd.read_csv('fifa_data.csv')
fifa
```

**HISTOGRAM**

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins = bins, color = '#EB4C2A')
plt.xticks(bins)
plt.ylabel('Numbers of Player')
plt.xlabel('Skill')
plt.title("Distribution of player skill", color = 'blue')
plt.show()
```

**PIE CHART**

```python
fifa.head(5)
```

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

labels = ['Left','Right']
color = ['#8D4F4F','#28828B']
plt.pie([left,right], labels = labels, colors = color, autopct = '%.2f')
plt.title('Preference Foot of players')
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
plt.style.use('ggplot')
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]
weights = [light,light_medium,medium,medium_heavy,heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)
plt.title('Weight of Professional Soccer Players (lbs)')
plt.pie(weights, labels = label, autopct='%.2f %%',pctdistance = 0.8, explode = explode)
plt.show()
```

**Box and Whiskers Chart**

```python
plt.style.use('default')
plt.figure(figsize=(5,8), dpi=100)
Barca = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
Real = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']
labels = ['Barca','Madrid','REVS']
bp = plt.boxplot([Barca,Real,revs],labels = labels,patch_artist=True, medianprops={'linewidth': 2})
plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    box.set(hatch = '/')
plt.show()
```
