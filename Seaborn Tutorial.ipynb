###IMPORT SEABORN

```python
import seaborn as sns
```

Get some datasets from Seaborn

```python
sns.get_dataset_names()
```

```python
diamond = sns.load_dataset('diamonds')
titanic = sns.load_dataset('titanic')
car = sns.load_dataset('car_crashes')
```

### Scatter Plot


```python
diamond
```

```python
# See the relationship between depth and price stored in diamonds
sns.scatterplot(x = 'depth', y='price', data = diamond, hue = 'color')
```

### Histogram and Distribution Plot

```python
sns.histplot(diamond['depth'], kde = True, bins = 39)
```

Easily see that the depth of diamond in between 60 to 65

```python
sns.displot(diamond['depth'], kde = True, bins = 39)
```

### Bar Plot


```python
ax = sns.barplot(x = 'cut', y = 'price', data = diamond)
# Add a for loop to give each bar an exact value
for bars in ax.containers:
  ax.bar_label(bars)
```

### Box plot and Strip plot

```python
sns.boxplot(x="cut",y='price', data = diamond)
```

```python
sns.stripplot(x = 'cut', y='price', data = diamond, dodge = True)
```

### Join plot

```python
sns.jointplot(x='depth', y='price', data = diamond)
```

### Heat map

```python
diamond.corr()
```

```python
sns.heatmap(diamond.corr(), annot = True, cmap = "YlGnBu")
```
